# Cargar dataset notas

```python
# Ruta principal donde están las carpetas con los CSV
ruta_principal = r"C:\Users\raulm\Documents\TFG\medicalReports\mimic-iv-note-deidentified-free-text-clinical-notes-2.2"

# Diccionario para almacenar los DataFrames de cada módulo
notes_data = {}

# Función para leer archivos comprimidos y almacenarlos en un diccionario
def cargar_datos(ruta_modulo, diccionario):
    for archivo in os.listdir(ruta_modulo):
        # Buscar archivos gz en la carpeta
        if archivo.endswith(".gz"):
            ruta_gz = os.path.join(ruta_modulo, archivo)
            nombre_tabla = archivo.replace(".csv.gz", "") 
            df = pd.read_csv(ruta_gz, compression="gzip", low_memory=False)
            diccionario[nombre_tabla] = df
            print(f"Cargado: {nombre_tabla} ({df.shape[0]} filas)")

# Cargar datos de notas
ruta_notes = os.path.join(ruta_principal, "note")
print ("Cargando datos de note...")
cargar_datos(ruta_notes, notes_data)
```

```python
df_notes = []
# Recorrer los archivos de notes que contengan subject_id y concatenarlos en un df nuevo
for archivo, df in notes_data.items():
    if 'subject_id' in df.columns:
        df_notes.append(df)
    else:
        print(f"Advertencia: {archivo} no tiene 'subject_id'.")


if df_notes:
    notes_data = pd.concat(df_notes, ignore_index=True)
    notes_data.to_csv("datasets/notes_dataset.csv", index=False)
    print("Dataset notes unificado guardado exitosamente.")
else:
    print("No se encontraron archivos válidos para unificar.")
```

## Extraer los subject_id en un fichero de texto

```python
# Cargar el dataset
df = pd.read_csv('/home/raulmartinez/medicalReports/datasets/dataset_notes_unificado.csv', low_memory=False) 

# Extraer todos los subject_id únicos
subject_ids = df['subject_id'].unique()

# Guardar los subject_id en un archivo de texto
with open('datasets/subject_ids.txt', 'w') as f:
    for subject_id in subject_ids:
        f.write(f"{subject_id}\n")

print(f"Se han guardado {len(subject_ids)} subject_id en 'subject_ids.txt'")
```
